## 車線数をカウント

```python
def count_lane(df):
    lane_cols = ['全車(1)', '全車(2)', '全車(追)','全車(路)']
    df.loc[:, 'lane_count'] = df.loc[:, lane_cols].isna().sum(axis=1)
    df.loc[:, 'lane_count'] = len(lane_cols) - df.loc[:, 'lane_count']
    return df

df = count_lane(df)
```
