
### 処理の流れ

- Author: 葛

- Input
    - /Input_original_data/traffic/TK_kannetsu_(20220501-20220621).CSV (943488, 41)
    - /Input_original_data/traffic/TK_kannetsu_(20220401-20220508).CSV (689472, 41)

    - ../Input_original_data/traffic/TK_kannetsu_(202104-202203).CSV (6622560, 41)

- 列【'全車(１)','全車(2)','全車(追)','全車(路)'】の欠損から車線数Lane_countを算出

['エラー数', '全車(2)', '大車(2)', '車線率(2)',

- それぞれOutputを
    - processed_taffic+'/{TARGET_ROAD}~20220621nonullcols.csv'
        - 関越サイズ(8110368, 41)

```python
import os
import networkx as nx
import pandas as pd
import missingno as msno
import pandas_profiling
import datetime
# import warnings
# warnings.simplefilter('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
# import japanize_matplotlib
plt.rcParams['font.family'] = 'Noto Sans JP'
```

```python
# data directory
PROCESSED_DATA_DIR = '../Input_processed_data'
ORI_DATA_DIR = '../Input_original_data'


# 道路名
TARGET_ROAD='tateyama'
#TARGET_ROAD='kannetsu'

TIMES={'(202104-202203)','(20220401-20220508)','(20220501-20220621)','(202206-09)'}
#TIMES={'(20220401-20220508)'}
# IC, 道路情報 csv
IC_CSV = f'{PROCESSED_DATA_DIR}/road_master/ic_merged.csv'
IC_NET_SUB_CSV = f'{PROCESSED_DATA_DIR}/road_master/tateyama_{TARGET_ROAD}_doronet_sub.csv'

# 検索量 csv
SEARCH_COUNT_DIR = f'{PROCESSED_DATA_DIR}/search_count'
SEARCH_COUNT_KANNETSU_CSV = f'{SEARCH_COUNT_DIR}/search-count_kannetsu.csv'


processed_taffic = f'{PROCESSED_DATA_DIR}/traffic'
```

```python
output_file=f'{processed_taffic}/{TARGET_ROAD}_20220930nonullcols.csv'
```

```python
for i, time in enumerate(TIMES):

    file_name=f'{ORI_DATA_DIR}/traffic/TK_{TARGET_ROAD}_{time}.CSV'
    
    df_tf= pd.read_csv(file_name,encoding='shift-jis', header=1)
    print(i,time,file_name, df_tf.shape)
    
    #'(20220401-20220508)'は時間軸に重複な部分があるため、20220501までを取り出す
    if time=='(20220401-20220508)':
        print('data of ', time)
        end_time=datetime.datetime(2022,5,1)
        
        dates = df_tf.apply(lambda r: datetime.date(r['年'], r['月'], r['日']), axis=1)
        times = df_tf['時刻'].apply(lambda t: datetime.datetime.strptime(t, '%H:%M').time())
        df_tf['datetime'] = [datetime.datetime.combine(d, t) for d, t in zip(dates, times)]

        df_tf=df_tf.loc[(df_tf['datetime']<end_time)]
        
        df_tf.drop(['datetime'],axis=1,inplace=True)
        print('after taking data of ~ 2022_5_01',df_tf.shape)
    
    #'(20220501-20220621)'は時間軸に重複な部分があるため、20220601までを取り出す
    if time=='(20220501-20220621)':
        print('data of ', time)
        end_time=datetime.datetime(2022,6,1)
        
        dates = df_tf.apply(lambda r: datetime.date(r['年'], r['月'], r['日']), axis=1)
        times = df_tf['時刻'].apply(lambda t: datetime.datetime.strptime(t, '%H:%M').time())
        df_tf['datetime'] = [datetime.datetime.combine(d, t) for d, t in zip(dates, times)]

        df_tf=df_tf.loc[(df_tf['datetime']<end_time)]
        
        df_tf.drop(['datetime'],axis=1,inplace=True)
        print('after taking data of ~ 2022_5_01',df_tf.shape)
    
    if i==0:
        df_all=df_tf
    else:
        df_all=df_all.append(df_tf)
print('---total size',df_all.shape)
```

'全車(1)', '全車(2)', '全車(追)'列の欠損から、車線数を算出

```python
def countLane(df_ex):
    #test_pd=test_pd[['全車', '全車(1)', '全車(2)', '全車(追)']]
    lanes=['全車(1)', '全車(2)', '全車(追)','全車(路)']
    df_ex['lane_count']=df_ex[lanes].isna().sum(axis=1)
    df_ex['lane_count']=len(lanes)-df_ex['lane_count']
    return df_ex
df_all=countLane(df_all)
```

```python
df_all.drop(['エラー数', '全車(2)', '大車(2)', '車線率(2)',
       'OCC(2)', '速度(2)', 'エラー数.1', 'エラー数.2', '全車(路)', '大車(路)', '車線率(路)', 'OCC(路)', '速度(路)',
       'エラー数.3'],inplace=True,axis=1)
```

```python
msno.matrix(df_all, labels=True)
```
