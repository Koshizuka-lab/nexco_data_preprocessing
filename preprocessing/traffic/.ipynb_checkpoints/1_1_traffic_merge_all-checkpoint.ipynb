## トラカンデータと県コード, 経緯度,天気情報, 制限速度をマージ¶

```python
def kukan2code(df_ex, df_ic):
    #name => code
    name2code = dict(zip(df_ic['ic_name'], df_ic['ic_code']))
    name2code2 = dict(zip(df_ic['dorapura_name'], df_ic['ic_code']))
    name2code.update(name2code2)

    # code => name
    code2name = dict(zip(df_ic['ic_code'], df_ic['ic_name']))

    # code => degree
    code2degree = dict(zip(df_ic['ic_code'], df_ic['degree']))
    
    #'\u3000' の文字化けを修正
    res = df_ex.assign(
        segment_name=(df_ex.区間名称
                      .str.replace('\u3000', '').str.replace('\u3000', '').str.split('〜'))
    )
    res = res.assign(
        start_name=res.segment_name.to_pandas().apply(lambda l: l[0]),
        end_name=res.segment_name.to_pandas().apply(lambda l: l[1])
    )

    # 新たに追加した処理: 上り・下りに応じてstart_name, end_nameを入れ替える
    up_start = res.loc[res.方向 == '上り', 'start_name']
    up_end = res.loc[res.方向 == '上り', 'end_name']

    res.loc[res.方向 == '上り', 'start_name'] = up_end
    res.loc[res.方向 == '上り', 'end_name'] = up_start

    res = (res
           .assign(direction=res.方向.map({'上り': 'up', '下り': 'down'}))
           .assign(
               start_code=res.start_name.map(name2code),
               end_code=res.end_name.map(name2code)))

    res = (res
           .assign(
               start_name=res.start_code.map(code2name),
               end_name=res.end_code.map(code2name))
           .assign(
               start_degree=res.start_code.map(code2degree),
               end_degree=res.end_code.map(code2degree))
           .pipe(lambda _df: _df.assign(degree_sum=_df.start_degree + _df.end_degree)))

    return res


def to_datetime(df):
    '''
    str型の「年」「月」「日」「時刻」 カラムからdatetime型の「datetime」カラムを作成
    '''
    res = df.assign(
        datetime=(
            df.年 + '/' + df.月 + '/' + df.日 + ' ' + df.時刻
        ).to_pandas().apply(lambda t: dt.datetime.strptime(t, '%Y/%m/%d %H:%M'))
    )
    return res


def is_holiday(weekday, date):
    if weekday >= 6 or jpholiday.is_holiday(date):
        return 1
    return 0


def fix_holiday(df):
    '''
    jpholidayライブラリを使って休日判定
    '''
    res = df.assign(
        is_holiday=(df.loc[:, ['dayofweek', 'datetime']]
                    .to_pandas().apply(lambda r: is_holiday(r.dayofweek, r.datetime), axis=1)
                   )
    )
    return res


def process_datetime(df):
    # ['年','月','日','時刻']からdatetime型の'datetime'カラムを作成
    res = to_datetime(df)

    # month, day, hour, minute情報を付加
    res = (res
           .assign(
               month=res.datetime.dt.month,
               day=res.datetime.dt.day,
               hour=res.datetime.dt.hour,
               minute=res.datetime.dt.minute,
               dayofweek=res.datetime.dt.dayofweek + 1))

    # is_holiday: {0, 1}
    res = fix_holiday(res)

    return res


def code2KP(df_ic, df_ex):
    code2KP = dict(zip(df_ic['ic_code'], df_ic['KP']))
    
    res = (df_ex
           .assign(
               start_KP=df_ex.start_code.map(code2KP),
               end_KP=df_ex.end_code.map(code2KP))
           .pipe(lambda _df: _df.assign(
               KP_distance=np.abs(_df.start_KP - _df.end_KP)))
          )
    
    return res
    

def code2pref_code(df_ic, df_ex):
    code2prefcode = dict(zip(df_ic['ic_code'],df_ic['pref_code']))
    
    res = df_ex.assign(
        start_pref_code=df_ex.start_code.map(code2prefcode),
        end_pref_code=df_ex.end_code.map(code2prefcode))
    
    return res

def code2lat_lon(df_ic, df_ex):
    code2lat = dict(zip(df_ic['ic_code'], df_ic['lat']))
    code2lng = dict(zip(df_ic['ic_code'], df_ic['lng']))
    
    res = df_ex.assign(
        start_lat=df_ex.start_code.map(code2lat),
        start_lng=df_ex.start_code.map(code2lng),
        end_lat=df_ex.end_code.map(code2lat),
        end_lng=df_ex.end_code.map(code2lng))
    
    return res


def mergeLimitSpeed(df_icnet, df_ex):
    res = df_ex.merge(df_icnet.loc[:, ['start_code', 'end_code', 'limit']],
                      on=['start_code','end_code'], how='left')
    
    # 館山道 制限速度 全ては100
    if TARGET_ROAD=='tateyama':
        res.limit.fillna(100, inplace=True)
        
    return res


def mergeWeather(weather_pd, df_ex):
    # '気圧（現地）', '降水量', '気温', '湿度', '日照時間','平均風速', '降雪量', '最深積雪量', '天気概況-昼', 'エリア',
    #'datetime', 'pref_code'
    #区間のstart県コードの天気情報マージ
    res = (df_ex
           .assign(date=df_ex.datetime.to_pandas().dt.date.astype(str))
           .merge(weather_pd, left_on=['start_pref_code', 'date'], right_on=['pref_code', 'date'], how='left')
           .drop(['エリア', 'pref_code'], axis=1))
    
    return res


def rename_cols(df):
    rename_map = {
        '全車': "allCars",
        '速度': 'speed',
        '気温': 'temperature',
        '降水量': 'rainfall',
        '日照時間': 'daylight_hours',
        '降雪量': 'snowfall',
        '平均風速': 'wind_speed',
        '天気概況-昼': 'weather_description',
        '最深積雪量': 'deepest_snowfall',
        '湿度': 'humidity',
        '気圧（現地）': 'pressure',
        'limit': 'limit_speed',
    }

    res = df.rename(columns=rename_map)
    return res
```

```python
def preprocess_traffic(df_traffic, df_ic, df_weather):
    #事前にカラムを限定して前処理を軽くする
    selected_cols = ['年', '月', '日', '時刻', '方向', 'KP', '区間名称', '全車', 'OCC', '速度','lane_count']
    res = df_traffic.loc[:, selected_cols]

    res = kukan2code(res, df_ic)
    res = process_datetime(res)
    res = code2pref_code(df_ic, res)
    res = code2lat_lon(df_ic, res)
    res = code2KP(df_ic, res)
    
    selected_cols = [
        'datetime', 'start_name', 'end_name', 'start_code', 'end_code', 
        'start_pref_code', 'end_pref_code', 'start_lat', 'end_lat', 'start_lng', 'end_lng',
        'start_degree', 'end_degree', 'month', 'day', 'dayofweek', 'is_holiday', 'hour', 'minute',
        'lane_count','KP_distance', 'KP', 'start_KP', 'end_KP', 'direction', 'OCC', '全車', '速度'
    ]
    res = res.loc[:, selected_cols]

    res = mergeWeather(df_weather, res)
    res = mergeLimitSpeed(df_icnet, res)
    
    # 列名変更
    res = rename_cols(res)
    return res
```

```python
df_ic = cudf.from_pandas(pd.read_csv(IC_CSV))
df_icnet = cudf.from_pandas(pd.read_csv(IC_NET_SUB_CSV, dtype={'start_code': str, 'end_code': str, 'road_code': str}))

df_weather = cudf.from_pandas(pd.read_csv(WEATHER_CSV))
df_weather.head(3)
```

```python
dtype = {
    '年': str, '月': str, '日': str
}

df_traffic = cudf.from_pandas(pd.read_csv(TRAFFIC_CSV, encoding='shift jis', dtype=dtype))
# df_traffic.drop('index', axis=1, inplace=True)
print(df_traffic.shape)
df_traffic.head(3)
```

```python
%time df_traffic_preprocessed = preprocess_traffic(df_traffic, df_ic.to_pandas(), df_weather)
df_traffic_preprocessed.tail()
```

```python
msno.matrix(df_traffic_preprocessed.to_pandas(), labels=True)
plt.show()
```

## 検索データとマージ

### 時間指定ありの検索データをマージ

```python
df_traffic = cudf.from_pandas(pd.read_pickle(f'./traffic_preprocessed_{TARGET_ROAD}.pkl'))
```

```python
search_all = pd.read_csv(SEARCH_COUNT_CSV, parse_dates=True, index_col='passing_time',
                        dtype={'start_code': str, 'end_code': str}).reset_index()
print(search_all.shape)
search_all.head(3)
```

```python
search_all.tail()
```

```python
search_all = cudf.from_pandas(search_all)
```

```python
def merge_search(df_ex, df_search):
    def fillna_in_group(g):
        col = g.filter(like='search').columns
        if not g[col].isna().all().all():
            g[col] = g[col].fillna(0)
        return g

    res = (df_ex
           .merge(df_search, how='left',
                  left_on=['datetime', 'start_code', 'end_code'], 
                  right_on=['passing_time', 'start_code', 'end_code'])
           .drop('passing_time', axis=1)
           .to_pandas()
           .groupby(['start_code', 'end_code'])
           .apply(fillna_in_group))
           
    res = cudf.from_pandas(res)
    return res
```

```python
msno.matrix(search_all.to_pandas(), labels=True)
plt.show()
```

```python
%time df_traffic = merge_search(df_traffic, search_all)
msno.matrix(df_traffic.to_pandas(), labels=True)
plt.show()
```

### 時間指定なしの検索データをマージ

```python
def merge_search_un(df_ex, df_search_un):
    def fillna_in_group(g):
        cols = g.filter(like='search_unspec').columns
        g[cols] = g[cols].fillna(method='ffill')
        return g

    res = (df_ex
           .merge(df_search_un, how='left',
                  left_on=['datetime', 'start_code', 'end_code'], 
                  right_on=['search_date', 'start_code', 'end_code'])
           .drop('search_date', axis=1)
           .to_pandas()
           .groupby(['start_code', 'end_code'])
           .apply(fillna_in_group)
           .reset_index(drop=True))
    
    res = cudf.from_pandas(res)
    return res
```

```python
search_unspec_all = pd.read_csv(SEARCH_COUNT_UNSPEC_CSV, parse_dates=True, index_col='search_date',
                                  dtype={'start_code': str, 'end_code': str}).reset_index()
print(search_unspec_all.shape, search_unspec_all.dtypes)
search_unspec_all.head(3)
```

```python
search_unspec_all = cudf.from_pandas(search_unspec_all)
```

```python
%time df_traffic = merge_search_un(df_traffic, search_unspec_all)
msno.matrix(df_traffic.to_pandas(), labels=True)
plt.show()
```
