```python
import missingno as msno
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.simplefilter('ignore')
import datetime
import seaborn as sns
import cudf
```

```python
# 道路名
# TARGET_ROAD='tateyama'
TARGET_ROAD='kannetsu'

# 交通量
PROCESSED_DATA_DIR = '.'
TRAFFIC_DIR = f'{PROCESSED_DATA_DIR}/traffic'
TRAFFIC_CSV = f'{TRAFFIC_DIR}/{TARGET_ROAD}_20220930all-merged_filled.csv'

OUTPUT_CSV = lambda sr: f'{TRAFFIC_DIR}/{TARGET_ROAD}_20220930all-merged_filled_{sr}.csv'
```

```python
def resample_traffic(df, sr):
    '''
    サンプリングレートsrでdfをリサンプル
    '''
    if sr in {'5min', '1h'}:
        AGG_COLS = ['allCars', 'speed', 'OCC']
        AGG_COLS_METHODS = {'allCars': 'sum', 'accum_speed': 'sum', 'OCC': 'mean'}
    else:
        AGG_COLS = ['allCars', 'speed', 'OCC', 'search_5min']
        AGG_COLS_METHODS = {'allCars': 'sum', 'accum_speed': 'sum', 'OCC': 'mean', 'search_5min': 'sum'}

    # リサンプリングしても値が変わらない列は間隔srで単純にサンプルしてくる
    df_st = cudf.from_pandas(
        df
        .drop(AGG_COLS, axis=1)
        .set_index('datetime')
        .groupby(['start_code', 'end_code'])
        .apply(lambda g: g.asfreq(sr))
        .reset_index('datetime')
        .reset_index(drop=True)
    )

    # 値を集計しながらリサンプリングする列（AGG_COLS）
    df_agg = cudf.from_pandas(
        df
        .assign(accum_speed=df['allCars'] * df['speed'])
        .set_index('datetime')
        .groupby(['start_code', 'end_code'])
        .apply(lambda g: g.resample(sr).agg(AGG_COLS_METHODS))
        .reset_index()
    )
    
    res = df_st.merge(df_agg, on=['datetime', 'start_code', 'end_code'])
    res = (res
           .assign(speed=res.accum_speed / (res.allCars + 1))
           .drop(['accum_speed'], axis=1)
           .loc[:, df.columns]) # 列の順番を保存
    
    if sr not in {'5min', '1h'}:
        res.rename(columns={'search_5min': f'search_{sr}'}, inplace=True)
    
    res = (res
           .sort_values(['start_code', 'end_code', 'datetime'])
           .reset_index(drop=True)
           .to_pandas())
    return res
```

```python
df_traffic_5min_o = pd.read_csv(TRAFFIC_CSV, parse_dates=True, index_col='datetime',
                dtype={'start_code': str, 'end_code': str, 'road_code': str}).reset_index()
print(df_traffic_5min_o.shape)
df_traffic_5min_o.head(3)
```

```python
df_traffic_5min = df_traffic_5min_o.copy()
df_traffic_5min.head(5)
```

## 欠損値を補間

```python
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)

    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)

    # Print some summary information
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
          " columns that have missing values.")

    # Return the dataframe with missing information
    return mis_val_table_ren_columns
```

```python
def col_interpolate(df, cols, method='linear'):
    """
    dfのcol列内の欠損を区間ごとに線形補間する

    Parameters
    ----------
    df: pandas.DataFrame
    col: str or List[str]
    """
    grouped_intp = lambda g: g.interpolate(method=method, axis=0)
    
    interpolated = df\
        .sort_values('datetime')\
        .groupby(['start_code', 'end_code'])[cols]\
        .apply(grouped_intp)\
        .sort_index()
    
    df_res = df.assign(**{
        col: interpolated[col]
        for col in cols
    })
    return df_res
```
