```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Noto Sans JP'
```

```python
# 道路名
# TARGET_ROAD = 'kannetsu'
TARGET_ROAD = 'touhoku'

# トラカンデータ
PERIOD = '20230817-20230930'
TRAFFIC_FILE = f'./traffic/raw_{TARGET_ROAD}_{PERIOD}_2KP.csv'
```

```python
# ! ls -1 | xargs -n1 -I@ sh -c 'iconv -f sjis -t utf-8 @ | tail -n +2 > new_@'
```

```python
df = pd.read_csv(TRAFFIC_FILE)
df.head()
```

```python
msno.matrix(df, labels=True)
plt.show()
```

## 車線数をカウント

```python
def count_lane(df):
    lane_cols = ['全車(1)', '全車(2)', '全車(追)','全車(路)']
    df.loc[:, 'lane_count'] = df.loc[:, lane_cols].isna().sum(axis=1)
    df.loc[:, 'lane_count'] = len(lane_cols) - df.loc[:, 'lane_count']
    return df

df = count_lane(df)
```

```python
df.lane_count.value_counts()
```

## エラーの多い列を削除

```python
DROP_COLS = [
    'エラー数', '全車(2)', '大車(2)', '車線率(2)', 'OCC(2)', '速度(2)', 'エラー数.1', 
    'エラー数.2', '全車(路)', '大車(路)', '車線率(路)', 'OCC(路)', '速度(路)','エラー数.3'
]

df.drop(DROP_COLS, axis=1, inplace=True)
```

```python
msno.matrix(df, labels=True)
plt.show()
```

全車線が使用できない場合、全車=0

```python
df.loc[df.lane_count == 0,  ['年', '月', '日', '時刻', '方向', '区間名称']]
```

## （第2回コンテスト用） 2kmピッチにトラカンが設置している区間を抽出
- 2kmピッチにトラカンが設置されている区間の集合は`traffic_counter_segments_2KP.pkl`に保存されている
- 2023/8/16以前のデータに関して、↑の区間に限定する

```python
import pickle

with open('traffic_counter_segments_2KP.pkl', 'rb') as f:
    tc_segments_dict = pickle.load(f)
```

```python
tc_segments_up = tc_segments_dict[TARGET_ROAD]['上り']
tc_segments_down = tc_segments_dict[TARGET_ROAD]['下り']
```

```python
cond_up = (df.区間名称.isin(tc_segments_up)) & (df.方向 == '上り')
cond_down = (df.区間名称.isin(tc_segments_down)) & (df.方向 == '下り')
```

```python
df_tmp = df.loc[cond_up | cond_down].reset_index(drop=True)
```

```python
# test
assert set(df_tmp.loc[df_tmp.方向 == '上り', '区間名称'].drop_duplicates()) == tc_segments_up
assert set(df_tmp.loc[df_tmp.方向 == '下り', '区間名称'].drop_duplicates()) == tc_segments_down
```

```python
df = df_tmp
```

## 年月日でソート -> 保存

```python
df.sort_values(['年','月','日'], inplace=True)
msno.matrix(df, labels=True)
plt.show()
```

```python
OUTPUT_FILE = f'./traffic/{TARGET_ROAD}_{PERIOD}_2KP.csv'
```

```python
%time df.to_csv(OUTPUT_FILE, index=False)
```
