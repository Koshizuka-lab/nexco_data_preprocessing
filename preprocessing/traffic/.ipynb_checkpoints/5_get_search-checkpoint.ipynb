## 時間指定なし

```python
search_unspec_cols = ['datetime', 'start_code', 'end_code','search_unspec_1d']
df_search_unspec = df.loc[:, search_unspec_cols]

df_search_unspec['date'] = df_search_unspec['datetime'].str[:10]
df_search_unspec.head()
```

```python
df_search_unspec = df_search_unspec.drop(['datetime'],axis=1)
```

```python
df_search_unspec = (df_search_unspec
                    .drop_duplicates(['start_code','end_code','date'])
                    .sort_values(['start_code','end_code']))

df_search_unspec = df_search_unspec.loc[:, ['date','start_code','end_code','search_unspec_1d']]
df_search_unspec.head()
```

```python
df_search_unspec.tail()
```

```python
msno.matrix(df_search_unspec, labels=True)
plt.show()
```

```python
df_search_unspec.dtypes
```

```python
fname_out = f'./search_count/search_unspec_data_20230801-20230816.csv'
df_search_unspec.to_csv(fname_out, index=False)
```
